2. Explain the concept of function arguments and parameters in Python.

Ans. Parameters are the names given to variables in a function definition. They act as placeholders for the values that will be passed into the function when it is called. Example: def greet(name) # 'name' is the parameter.

Arguments are the actual values that are passed to the function when it is called. Example: greet("anurag") # "anurag" is the argument.

7. What are the advantages of using generators vs regular functions?

Ans. Memory Efficiency: Generators use less memory as they generate items one at a time.

Lazy Evaluation: Values are produced on-the-fly, improving performance.

Simplified Code: Generator functions are often more readable and maintainable.

Infinite Sequences: Generators can handle infinite sequences without memory issues.
